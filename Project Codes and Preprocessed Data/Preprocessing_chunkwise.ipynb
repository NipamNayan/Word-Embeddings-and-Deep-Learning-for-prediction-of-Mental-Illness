```python
import os
import numpy as np
import pandas as pd
import pickle
from io import StringIO
import xml.etree.ElementTree as ET
```

```python
pos_folder=r"C:\Users\SONY\Desktop\WORK\New\train\positive_examples"
neg_folder=r"C:\Users\SONY\Desktop\WORK\New\train\negative_examples"
```

```python
listfiles=os.listdir(pos_folder)
listfiles2=os.listdir(neg_folder)
```

```python
print(listfiles)
```

```python
# stores the number of writings for each user
cor1 = []  
cor1a =[]  
cor1b =[] 
```

```python

for file in listfiles:
    filepath=os.path.join(pos_folder,file)  
    print(filepath)
    #print(os.listdir(filepath))
    list2 = os.listdir(filepath)
    for data in list2:
        tree = ET.parse(filepath+'\\'+ data)
        print(filepath+'\\'+data)
   
        
        root = tree.getroot() #prints ID
        print(root)
        pos_txt='' 
        for child in root:
            if child.tag!='ID':
                pos_txt= pos_txt+child[0].text+child[3].text 
        cor1.append(root[0].text+'\n'+pos_txt)
        cor1a.append(pos_txt)
        cor1b.append(root[0].text)
    #print(cor1a)
```

```python
Pos_Class = []
for i in cor1a:
    row = []
    row.append(i)
    row.append(1)
    Pos_Class.append(row)
```

```python
len(Pos_Class)
```

```python
cor2a = []
cor2 =[]
corb = []
```

```python
i=1
for file in listfiles2:
    filepath=os.path.join(neg_folder,file)
    #print(filepath)
    list_neg = os.listdir(filepath)
    for data in list_neg:
        i=i+1
        print(i)
        tree = ET.parse(filepath+'\\'+data)
        root = tree.getroot()
        neg_txt='' 
        for child in root:
            if child.tag!='ID':
                neg_txt= neg_txt+child[0].text+child[3].text
        cor2.append(root[0].text+'\n'+neg_txt) #prints subject id with text
        cor2a.append(neg_txt)
        corb.append(root[0].text)
```

```python
len(cor2a)
```

```python
Neg_Class = []
for i in cor2a:
    row = []
    row.append(i)
    row.append(0)
    Neg_Class.append(row)

corpus = cor1+cor2
type(corpus)
```

```python
Combined = Pos_Class + Neg_Class
Combined = pd.DataFrame(Combined,columns=['Text','Label'])
```

```python
Combined
```

```python
Subjects = cor1b+corb
Subjects = pd.DataFrame(Subjects,columns = ['Id'])
```

```python
Anorexia_train  = [Combined,Subjects]
Anorexia_train = pd.concat(Anorexia_train , axis=1)
```
